```python
### https://adventofcode.com/2023/day/14

import numpy as np
import hashlib


def read_file(file_name):
    with open(file_name, 'r') as f_in:
        return list(filter(None, [line.strip() for line in f_in.readlines()]))
    
    
def slide_2_north(matrix, num_rows, num_cols):
    for i_row in range(num_rows):
        for i_col in range(num_cols):
            if matrix[i_row, i_col] == 'O':
                row_iter = i_row
                while row_iter > 0 and matrix[row_iter - 1, i_col] == '.':
                    matrix[row_iter - 1, i_col] = 'O'
                    matrix[row_iter, i_col] = '.'
                    row_iter -= 1                    
    return matrix


def slide_2_west(matrix, num_rows, num_cols):
    for i_row in range(num_rows):
        for i_col in range(num_cols):
            if matrix[i_row, i_col] == 'O':
                col_iter = i_col
                while col_iter > 0 and matrix[i_row, col_iter - 1] == '.':
                    matrix[i_row, col_iter - 1] = 'O'
                    matrix[i_row, col_iter] = '.'
                    col_iter -= 1                    
    return matrix      


def slide_2_south(matrix, num_rows, num_cols):
    row_reverse = list(range(num_rows))[::-1]
    for i_row in row_reverse:
        for i_col in range(num_cols):
            if matrix[i_row, i_col] == 'O':
                row_iter = i_row
                while row_iter < num_rows - 1 and matrix[row_iter + 1, i_col] == '.':
                    matrix[row_iter + 1, i_col] = 'O'
                    matrix[row_iter, i_col] = '.'
                    row_iter += 1                    
    return matrix


def slide_2_east(matrix, num_rows, num_cols):
    col_reverse = list(range(num_cols))[::-1]
    for i_row in range(num_rows):
        for i_col in col_reverse:
            if matrix[i_row, i_col] == 'O':
                col_iter = i_col
                while col_iter < num_cols - 1 and matrix[i_row, col_iter + 1] == '.':                    
                    matrix[i_row, col_iter + 1] = 'O'
                    matrix[i_row, col_iter] = '.'
                    col_iter += 1                    
    return matrix   

            
def calc_load(matrix, num_rows): 
    row_factor = list(range(1, num_rows + 1))[::-1]
    row_idx_list = list(map(str, range(num_rows)))
    row_factor_by_index = dict(zip(row_idx_list, row_factor))
    sum_load = 0
    for i_row in range(num_rows):
        num_rocks = list(matrix[i_row, :]).count('O')  
        current_factor = int(row_factor_by_index[str(i_row)])
        sum_load += num_rocks * current_factor
    return sum_load


def get_solution_1(data):
    matrix = np.array([list(line) for line in data])
    num_rows, num_cols = matrix.shape
    matrix = slide_2_north(matrix, num_rows, num_cols)
    load = calc_load(matrix, num_rows)
    return load
        

def run_cycle(matrix, num_rows, num_cols):    
    matrix = slide_2_north(matrix, num_rows, num_cols)
#     print("North: \n", matrix)
    matrix = slide_2_west(matrix, num_rows, num_cols)
#     print("West: \n", matrix)
    matrix = slide_2_south(matrix, num_rows, num_cols)
#     print("South: \n", matrix)
    matrix = slide_2_east(matrix, num_rows, num_cols)
#     print("East: \n", matrix)
    return matrix
 
    
def get_solution_2(data, cycles):
    cycles = 1000000000    
    hash_results = []
    loads = []
    matrix = np.array([list(line) for line in data])
    num_rows, num_cols = matrix.shape
    for cycle in range(cycles):
#         print(f"### CYCLE {cycle}:")
        matrix = run_cycle(matrix, num_rows, num_cols)
#         print(matrix)
        matrix_str = '###'.join([''.join(list(matrix[i_row, :])) for i_row in range(num_rows)])
        hash_res = hashlib.sha256(matrix_str.encode()).digest()
        load_res = calc_load(matrix, num_rows)
        
        if hash_res in hash_results:            
            rep_seq_start = hash_results.index(hash_res)    
            load_idx = loads.index(load_res) if load_res in loads else -1
            if rep_seq_start == load_idx:            
                print(f"REPETITION found in cycle {cycle} with hash {hash_res}")
                rep_seq_end = cycle
                rep_seq_len, non_rep_seq_len = rep_seq_end - rep_seq_start, rep_seq_start
                remained_cycles = cycles - non_rep_seq_len
                additional_cycles = remained_cycles % rep_seq_len
                return loads[hash_results.index(hash_results[rep_seq_start - 1 + additional_cycles])]
            
        hash_results.append(hash_res)
        loads.append(load_res)




file = "data/day_14.txt"
# file = "data/day_14_test.txt"

num_cycles = 1000000000

data = read_file(file)

solution_1 = get_solution_1(data)
solution_2 = get_solution_2(data, num_cycles)

print("Day 14 - Solution 1:", solution_1)
print("Day 14 - Solution 2:", solution_2)
```

```python
### https://adventofcode.com/2023/day/15


def read_file(file_name):
    with open(file_name, 'r') as f_in:
        return list(filter(None, [line.strip() for line in f_in.readlines()]))
    

def get_hash(string):
    hash_val = 0
    for char in string:
        ascii_char = ord(char)
        hash_val = ((hash_val + ascii_char) * 17) % 256
    return hash_val
        
    
def parse_data(data):
    return list(filter(None, sum([line.strip().split(',') for line in data], [])))

    
def get_solution_1(data):
    hashes = []
    for item in data:
        hashes.append(get_hash(item))    
    return sum(hashes)

def get_hashmap(data):
    hashmap = dict()
    for item in data:
        chg_str, remove_str, k = "=", "-", -1
        to_remove = remove_str in item
        if chg_str in item:
            k, v = item.split(chg_str)            
        elif to_remove:
            k = item.replace(remove_str, '')
        box = get_hash(k)
        
        if box not in hashmap:
            hashmap[box] = []
            
        box_keys = [item["key"] for item in hashmap[box]]

        if not to_remove:
            if k in box_keys:
                hashmap[box] = [item if item["key"] != k else {"key": k, "value": v} for item in hashmap[box]]
            else:
                hashmap[box].append({"key": k, "value": v})
        else:
            if k in box_keys:
                hashmap[box] = [item for item in hashmap[box] if item["key"] != k]
    return hashmap


def calc_focusing_power(hashmap):
    res = 0
    for box_num, items in hashmap.items():
        for idx, lense in enumerate(items):
            res += (box_num + 1) * (idx + 1) * int(lense["value"])
    return res
    
def get_solution_2(data):
    hashmap = get_hashmap(data)
    return calc_focusing_power(hashmap)




file = "data/day_15.txt"
# file = "data/day_15_test.txt"

data = parse_data(read_file(file))
solution_1 = get_solution_1(data)
solution_2 = get_solution_2(data)

print("Day 15 - Solution 1:", solution_1)
print("Day 15 - Solution 2:", solution_2)
```

```python
### https://adventofcode.com/2023/day/18

import re
import numpy as np




def read_file(file_name):
    with open(file_name, 'r') as f_in:
        return list(filter(None, [line.strip() for line in f_in.readlines()]))


def parse_input(data):
    sum_dict = {"U": 0, "D": 0, "L": 0, "R": 0}
    dig_plan = [re.split(r'\s+', item) for item in data]
    for i, item in enumerate(dig_plan):
        val = int(item[1])
        dig_plan[i][1] = val
        sum_dict[item[0]] += val
        dig_plan[i][2] = re.sub("\(|\)|\s", "", item[2])
    return dig_plan, sum_dict


def get_dig_trench(dig_plan, directions_sum):
    direction_op = {"U": -1, "D": 1, "L": -1, "R": 1}
    num_rows = (directions_sum["U"] + directions_sum["D"]) * 2
    num_cols = (directions_sum["L"] + directions_sum["R"]) * 2
    matrix = np.array([['.'] * num_cols] * num_rows)
    current_pos = (directions_sum["U"] - 1, directions_sum["L"] - 1)
    for item in dig_plan:
        direction, val, _ = item
        val = val * direction_op[direction]
        row = current_pos[0]
        col = current_pos[1]
        matrix[row, col] = "#"
        if direction in ["U", "D"]:         
            r_shift = row + val
            r_min, r_max = min(row, r_shift + 1), max(row, r_shift + 1)
            for r in range(r_min, r_max):
                matrix[r, col] = "#"
            current_pos = (r_shift, col)
        if direction in ["L", "R"]:
            c_shift = col + val
            c_min, c_max = min(col, c_shift + 1), max(col, c_shift + 1)
            for c in range(c_min, c_max):
                matrix[row, c] = "#"
            current_pos = (row, c_shift)               
    return matrix
        

def get_reduced_matrix(matrix):
    num_row, num_col = matrix.shape
    print(num_row, num_col)
    
    min_row_idx = min([list(matrix[i_row, :]).index('#') for i_row in range(num_row) if '#' in list(matrix[i_row, :])])
    max_row_idx = 1 + max([num_row - list(matrix[i_row, :])[::-1].index('#') for i_row in range(num_row) if '#' in list(matrix[i_row, :])[::-1]])
    min_col_idx = min([list(matrix[:, i_col]).index('#') for i_col in range(num_col) if '#' in list(matrix[:, i_col])])
    max_col_idx = 1 + max([num_col - list(matrix[:, i_col])[::-1].index('#') for i_col in range(num_col) if '#' in list(matrix[:, i_col])[::-1]])
    
    print(min_row_idx, max_row_idx, min_col_idx, max_col_idx)
    matrix_reduced = np.array(matrix[min_col_idx : max_col_idx + 1, min_row_idx : max_row_idx + 1]    )
    return matrix_reduced



def all_have_border(matrix, i_row, i_col):
    up = "#" in matrix[:i_row, i_col]
    down = "#" in matrix[i_row:, i_col]
    left = "#" in matrix[i_row, :i_col]
    right = "#" in matrix[i_row, i_col:]
    return all([up, down, left, right])


# def get_reduced_matrix(matrix):
#     num_row, num_col = matrix.shape
#     print(num_row, num_col)
#     for i_row in range(num_row):
#         for i_col in range(num_col):
#             if matrix[i_row, i_col] == ".":
#                 up = "#" in matrix[:i_row, i_col]
#                 down = "#" in matrix[i_row:, i_col]
#                 left = "#" in matrix[i_row, :i_col]
#                 right = "#" in matrix[i_row, i_col:]
#                 if all_have_border(matrix, i_row, i_col):
#                     i_up = matrix[:i_row, i_col].index('#')
#                     i_down = matrix[i_row:, i_col].index('#')
#                     i_left = matrix[i_row, :i_col].index('#')
#                     i_right = matrix[i_row, i_col:].index('#')
#                     up = all([all_have_border(matrix, i, i_col) for i in range(i_up, i_row)])
#                     down = all([all_have_border(matrix, i, i_col) for i in range(i_row, i_down)])
#                     left = all([all_have_border(matrix, i_row, i) for i in range(i_left, i_col)])
#                     left = all([all_have_border(matrix, i_row, i) for i in range(i_col, i_right)])
#                     if all([up, down, left, right]):
#                         matrix[i_row, i_col] = "#"
                    
    
    
def fill_lagoon(matrix):
    pass
#     num_row, num_col = matrix.shape
    
#     for i_row in range(num_row):      
# #         if ''.join(set(matrix[i_row,:])) != ".":
#         for i_col in range(num_col):
# #                 if ''.join(set(matrix[:, i_col])) != ".":
#             if matrix[i_row, i_col] == ".":
#                 up = "#" in matrix[:i_row, i_col]
#                 down = "#" in matrix[i_row:, i_col]
#                 left = "#" in matrix[i_row, :i_col]
#                 right = "#" in matrix[i_row, i_col:]
#                 if all([up, down, left, right]):
#                     matrix[i_row, i_col] = "#"
#     return matrix

        
        
def get_solution_1(matrix):
    volume = 0
    for row in matrix:
        volume += ''.join(row).count("#")
    return volume

    
def get_solution_2(data):
    pass




file = "data/day_18.txt"
# file = "data/day_18_test.txt"
out_matrix = "data/day_18_matrix.txt"
out_matrix_filled = "data/day_18_matrix_filled.txt"

data, sum_directions = parse_input(read_file(file))
dig_trench = get_dig_trench(data, sum_directions)
# print('\n'.join([''.join(list(row)) for row in dig_trench]))
red_matrix = get_reduced_matrix(dig_trench)

with open(out_matrix, 'w') as f_out:
    f_out.write('\n'.join([''.join(list(row)) + '\n' for row in red_matrix]))
# print('\n'.join([''.join(list(row)) for row in red_matrix]))
    
lagoon = fill_lagoon(red_matrix)
# print('\n'.join([''.join(list(row)) for row in lagoon]))

with open(out_matrix_filled, 'w') as f_out:
    f_out.write('\n'.join([''.join(list(row)) + '\n' for row in lagoon]))

solution_1 = get_solution_1(lagoon)
solution_2 = get_solution_2(data)

print("Day 18 - Solution 1:", solution_1) # to high: 54980
print("Day 18 - Solution 2:", solution_2)
```
